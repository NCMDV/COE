```python
import numpy as np # linear algebra
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from keras.preprocessing.text import Tokenizer
from keras.preprocessing import sequence
import keras
import tensorflow as tf
```

```python
df = pd.read_csv('QAs_Category_SubCategory_Servicenow.csv')
df.head()
```

```python
df.drop(['Answer','Subcategory'], axis=1, inplace=True)
df
```

```python
df['Category'].value_counts()
```

```python
df.info
```

```python
Category_n = LabelEncoder()
df['Category_n'] = Category_n.fit_transform(df['Category'])
df
```

```python
df['Category_n'].value_counts()
```

```python
x = df['Question']
y = df['Category_n']
y = y.to_numpy()
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2)
```

```python
x_train.shape , y_train.shape, x_test.shape, y_test.shape
```

```python
max_words = 1100
max_len = 200



tok = Tokenizer(num_words=max_words) 
tok.fit_on_texts(x_train)            
seq = tok.texts_to_sequences(x_train) 
seq_matrix = sequence.pad_sequences(sequences=seq, maxlen=max_len)   

seq_matrix
```

```python
test_seq = tok.texts_to_sequences(x_test)
test_seq_matrix = sequence.pad_sequences(test_seq, maxlen=max_len)

test_seq_matrix
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Embedding(input_dim=max_words, output_dim=6, input_length=max_len))
model.add(tf.keras.layers.LSTM(64, activation='tanh'))
model.add(tf.keras.layers.Dense(220, activation='relu'))
model.add(tf.keras.layers.Dropout(0.5))
model.add(tf.keras.layers.Dense(6, activation='softmax'))  # 6 units for classification

model.compile(optimizer=tf.keras.optimizers.AdamW(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
history = model.fit(seq_matrix, y_train, batch_size=2, epochs=100, validation_split=0.2)
```

```python
model.save('./Question_category_classification.h5')
print("Model saved successfully!")

loaded_model = tf.keras.models.load_model('./Question_category_classification.h5')
print("Model loaded successfully!")
```
